```python
dictt = {}
a = df_new[df_new.buyer_country == df_new.seller_country].groupby('brand').total.mean().index
b = df_new[df_new.buyer_country == df_new.seller_country].groupby('brand').total.mean().values
res = list(zip(a,b))
for i in res:
    dictt[i[0]] = i[1]
sorted(dictt.items(), key=lambda x: x[1], reverse=True)
```

```python
dictt2 = {}
a = df_new[df_new.buyer_country != df_new.seller_country].groupby('brand').total.mean().index
b = df_new[df_new.buyer_country != df_new.seller_country].groupby('brand').total.mean().values
res = list(zip(a,b))
for i in res:
    dictt2[i[0]] = i[1]
sorted(dictt2.items(), key=lambda x: x[1], reverse=True)
```

```python
dictt3 = {}
for k,v in dictt.items():
    for c,d in dictt2.items():
        if k == c:
            t = abs(v - d)
            dictt3[k] = t
sorted(dictt3.items(), key=lambda x: x[1], reverse=True)
```
